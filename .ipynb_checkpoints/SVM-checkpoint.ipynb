```python
from sklearn import datasets
import numpy as np
iris = datasets.load_iris()
X= iris.data[: , [0,3]]
y= iris.target
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test , y_train , y_test = train_test_split(X , y , 
                                                      test_size = 0.3 , random_state=1 , stratify = y) #Qui sto facendo lo split dei dataset di test e train col 30%, ho assgnato le 4 variabili
from sklearn.preprocessing import StandardScaler #faccio un preprocessing del dataset facendo una feature standardization
sc = StandardScaler()
sc.fit(X_train)
X_train_std = sc.transform(X_train)
X_test_std = sc.transform(X_test)
```

```python
from sklearn.svm import SVC
svm = SVC(kernel ='linear' , C=1.0 , random_state=0)
svm.fit(X_train_std , y_train)
```

```python
import funzioni_utili as fu
```

```python
from matplotlib.colors import ListedColormap
import matplotlib.pyplot as plt
X_combined_std = np.vstack((X_train_std , X_test_std))
y_combined = np.hstack((y_train , y_test))
fu.plot_dec_reg(X_combined_std , y_combined , classifier=svm, test_idx=range(105,150))
plt.xlabel('Feature 3')
plt.ylabel('Feature 4')
plt.legend(loc='upper left')
plt.show()
```

```python
y_pred = svm.predict(X_test_std)
print('Errori: %d' % (y_test != y_pred).sum())
from sklearn.metrics import accuracy_score
print('Accuratezza %.2f' % accuracy_score(y_test , y_pred))
```
